Definiamo ora una funzione per calcolare centro $C(x_c,y_c)$ e raggio $r$ di una circonferenza
$$x^2 + y^2 +ax +by +c = 0$$

$$x_c = -a/2, y_c = -b/2, r = \sqrt {x_c^2 + y_c^2 -c}$$

```python
# Definiamo ora una funzione che calcola centro e raggio di una circonferenza
def circonferenza(a,b,c):
    x_c = -a/2
    y_c = -b/2
    r = (x_c**2 + y_c**2 - c)**0.5
    return x_c, y_c, r
```

Definiamo ora una circonferenza $$x^2 + y^2 +4x +8y -10 = 0$$

```python
a= 4
b= 8
c=-10
x_c, y_c, r = circonferenza(a,b,c)
print(f"Il centro della circonferenza è ({x_c},{y_c})")
print(f"Il raggio della circonferenza è {r}")
```

**ESERCIZIO**
Dato un punto del piano $P(x_0,y_0)$ e una retta $s: ax+by+c=0$, scrivi una funzione per calcolare la distanza del punto dalla retta.
Ricorda $$d(P,s)=\frac {|a*x_0+b*y_0+c|}{\sqrt{a^2+b^2}} $$

La funzione deve essere chiamata con
d = distanza(x0,y0,a,b,c)

```python
import math

def distanza(x0, y0, a, b, c):
    """Calcola la distanza di un punto (x0, y0) da una retta ax + by + c = 0."""
    numeratore = abs(a * x0 + b * y0 + c)
    denominatore = math.sqrt(a**2 + b**2)
    if denominatore == 0:
        return float('inf') # La retta non è definita (a=0, b=0)
    return numeratore / denominatore
```

### Esempio di utilizzo della funzione `distanza`

```python
# Punto P(x0, y0)
x0_test = 1
y0_test = 2

# Retta ax + by + c = 0
a_test = 3
b_test = 4
c_test = -5

d_calcolata = distanza(x0_test, y0_test, a_test, b_test, c_test)
print(f"La distanza del punto ({x0_test}, {y0_test}) dalla retta {a_test}x + {b_test}y + {c_test} = 0 è: {d_calcolata}")
```

```python
def fibonacci(n):
    """Genera una lista con i primi n numeri della successione di Fibonacci."""
    sequenza = []
    a, b = 0, 1

    for _ in range(n):
        sequenza.append(a)
        # Aggiornamento simultaneo: a diventa b, b diventa la somma dei due
        a, b = b, a + b

    return sequenza

# Esempio: generiamo i primi 10 numeri
n_elementi = 15
risultato = fibonacci(n_elementi)
print(f"I primi {n_elementi} numeri di Fibonacci sono: {risultato}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plotta_funzione(f, x_min, x_max, punti=400, titolo="Grafico della funzione"):
    """
    Riceve una funzione f e la plotta nell'intervallo [x_min, x_max].
    """
    # Creiamo un array di valori x nell'intervallo scelto
    x = np.linspace(x_min, x_max, punti)

    # Calcoliamo i valori y corrispondenti
    # Usiamo np.vectorize per assicurarci che funzioni anche con funzioni non native NumPy
    f_vectorized = np.vectorize(f)
    y = f_vectorized(x)

    # Creazione del grafico
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label=f"f(x)", color="blue", linewidth=2)

    # Aggiunta di dettagli estetici
    plt.title(titolo)
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.axhline(0, color='black', linewidth=1) # Asse X
    plt.axvline(0, color='black', linewidth=1) # Asse Y
    plt.legend()

    # Mostra il grafico
    plt.show()

# --- ESEMPI DI UTILIZZO ---

# 1. Usando una funzione lambda (es: parabola)
plotta_funzione(lambda x: x**2 - 4, -5, 5, titolo="Parabola: x² - 4")

# 2. Usando una funzione matematica (es: seno)
plotta_funzione(np.sin, -2 * np.pi, 2 * np.pi, titolo="Funzione Seno")
```

We can also use variables inside the [] operator

```python
a = 10
b="Hello"
squares=[1,4,9,16,25,36]
list2 = [a,b,squares]   # i can insert into list
print(list2)
```

```python
# Lists can be indexed and sliced:
print(squares)
print(squares[0])  # indexing returns the item
print(squares[0:3]) # slicing returns a new list
print(squares[-1]) # start from the last element
print(squares[1:-1]) # print without the first and the last element

# All slice operations return a new list containing the requested elements.
```

```python
# Lists also support operations like concatenation:
print(squares + [36, 49, 64, 81, 100])
```

```python
# Unlike strings, which are immutable, lists are a mutable type, i.e. it
# is possible to change their content:
cubes = [1, 8, 27, 65, 125]  # something's wrong here, the cube of 4 is 64!
cubes[3] = 64  # replace the wrong value
print( cubes )
```
