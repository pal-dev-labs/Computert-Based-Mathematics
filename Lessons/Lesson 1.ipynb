Definiamo ora una funzione per calcolare centro $C(x_c,y_c)$ e raggio $r$ di una circonferenza
$$x^2 + y^2 +ax +by +c = 0$$

$$x_c = -a/2, y_c = -b/2, r = \sqrt {x_c^2 + y_c^2 -c}$$

```python
# Definiamo ora una funzione che calcola centro e raggio di una circonferenza
def circonferenza(a,b,c):
    x_c = -a/2
    y_c = -b/2
    r = (x_c**2 + y_c**2 - c)**0.5
    return x_c, y_c, r
```

Dato un punto del piano $P(x_0,y_0)$ e una retta $s: ax+by+c=0$, scrivi una funzione per calcolare la distanza del punto dalla retta.
Ricorda $$d(P,s)=\frac {|a*x_0+b*y_0+c|}{\sqrt{a^2+b^2}} $$

La funzione deve essere chiamata con
d = distanza(x0,y0,a,b,c)

SUGGERIMENTO: modulo di un numero: $|-3|$ si implementa con la funzione **abs(-3)**

```python
# import math
def distanza(x0, y0, a, b, c):
    """Calcola la distanza di un punto (x0, y0) da una retta ax + by + c = 0."""
    numeratore = abs(a * x0 + b * y0 + c)
    denominatore = (a**2 + b**2)**0.5
    if denominatore == 0:
        return float('inf') # La retta non è definita (a=0, b=0)
    return numeratore / denominatore
```

Data una parabola con asse parallelo all'asse y, di equazione $$y = ax^2 + bx +c $$ scrivi una funzione per calcolare le coordinate del Vertice $V(x_v,y_v)$, del Fuoco $F(x_f,y_f)$ e l'equazione della Direttrice $y=k$

Ricorda che
$$ V = (-\frac{b}{2a}, - \frac{b^2-4ac}{4a}) $$
$$ F = (-\frac{b}{2a}, - \frac{Δ}{4a}+\frac{1}{4a})  $$
$$ y = - \frac{Δ}{4a} - \frac{1}{4a} $$

la chiamata alla funzione deve essere

x_v,y_v,x_f,y_f,k = parabola(a,b,c)

Apri gemini e digita nel prompt "scrivi il codice python per ", poi incolla la richiesta dell'esercizio


```python
def parabola(a, b, c):
    """Calcola le coordinate del Vertice, del Fuoco e l'equazione della Direttrice
    per una parabola y = ax^2 + bx + c.
    """
    delta = b**2 - 4*a*c

    # Vertice
    x_v = -b / (2*a)
    y_v = -delta / (4*a)

    # Fuoco
    x_f = -b / (2*a)
    y_f = (-delta / (4*a)) + (1 / (4*a))

    # Direttrice
    k = (-delta / (4*a)) - (1 / (4*a))

    return x_v, y_v, x_f, y_f, k
```

```python
# Esempio di utilizzo della funzione parabola
a_test = 1
b_test = -4
c_test = 3

x_v, y_v, x_f, y_f, k = parabola(a_test, b_test, c_test)

print(f"Per la parabola y = {a_test}x^2 + {b_test}x + {c_test}:")
print(f"Vertice V: ({x_v}, {y_v})")
print(f"Fuoco F: ({x_f}, {y_f})")
print(f"Equazione della Direttrice: y = {k}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, FloatSlider

# Assicurati che la funzione 'parabola' sia definita nel tuo notebook
def parabola(a, b, c):
    delta = b**2 - 4*a*c
    x_v = -b / (2*a)
    y_v = -delta / (4*a)
    x_f = x_v
    y_f = (-delta / (4*a)) + (1 / (4*a))
    k = (-delta / (4*a)) - (1 / (4*a))
    return x_v, y_v, x_f, y_f, k

def plot_interactive_parabola(a, b, c):
    # Evitiamo la divisione per zero se a è 0 (non sarebbe una parabola)
    if a == 0:
        print("Il coefficiente 'a' deve essere diverso da 0 per avere una parabola.")
        return

    x_v, y_v, x_f, y_f, k = parabola(a, b, c)

    # Genera i punti
    x = np.linspace(x_v - 10, x_v + 10, 400)
    y = a * x**2 + b * x + c

    plt.figure(figsize=(10, 8))
    plt.plot(x, y, label=f'$y = {a:.2f}x^2 + {b:.2f}x + {c:.2f}$', color='dodgerblue', linewidth=2)

    # Vertice e Fuoco
    plt.scatter(x_v, y_v, color='red', s=100, zorder=5, label=f'Vertice V({x_v:.2f}, {y_v:.2f})')
    plt.scatter(x_f, y_f, color='green', s=100, zorder=5, label=f'Fuoco F({x_f:.2f}, {y_f:.2f})')

    # Direttrice
    plt.axhline(y=k, color='purple', linestyle='--', linewidth=1.5, label=f'Direttrice $y = {k:.2f}$')

    # Impostazioni grafico
    plt.title('Analisi Interattiva della Parabola', fontsize=14)
    plt.axvline(0, color='black', linewidth=0.8)
    plt.axhline(0, color='black', linewidth=0.8)
    plt.grid(True, linestyle=':', alpha=0.6)
    plt.legend(loc='upper right')

    # Mantieni il grafico centrato e proporzionato
    plt.ylim(y_v - 5, y_v + 15 if a > 0 else y_v + 5)
    plt.xlim(x_v - 10, x_v + 10)
    plt.show()

# Creazione degli slider
interact(plot_interactive_parabola,
         a=FloatSlider(min=-5.0, max=5.0, step=0.1, value=1.0, description='a:'),
         b=FloatSlider(min=-10.0, max=10.0, step=0.1, value=0.0, description='b:'),
         c=FloatSlider(min=-10.0, max=10.0, step=0.1, value=0.0, description='c:'))
```

Crea una funzione in python che simula il proiettile di un cannone. Specificata la componente x e y della velocità iniziale, la massa della palla, la funzione plotta la traiettoria e calcola la gittata. 

```python
import numpy as np
import matplotlib.pyplot as plt

def simula_cannone(v0x, v0y, m, k=0.1):
    """
    Simula la traiettoria di un proiettile con resistenza dell'aria.
    v0x, v0y: componenti della velocità iniziale (m/s)
    m: massa del proiettile (kg)
    k: coefficiente di attrito viscoso (kg/s)
    """
    g = 9.81  # Accelerazione di gravità (m/s^2)
    dt = 0.01 # Passo temporale
    
    # Inizializzazione liste
    x = [0.0]
    y = [0.0]
    vx = v0x
    vy = v0y
    
    # Ciclo di simulazione (finché il proiettile non tocca terra)
    while True:
        # Calcolo delle accelerazioni (F = ma -> a = F/m)
        # Forza peso: -m*g (solo su y)
        # Forza di attrito: -k*v
        ax = -(k * vx) / m
        ay = -g - (k * vy) / m
        
        # Aggiornamento velocità (Eulero)
        vx += ax * dt
        vy += ay * dt
        
        # Aggiornamento posizione
        next_x = x[-1] + vx * dt
        next_y = y[-1] + vy * dt
        
        if next_y < 0: # Tocca terra
            # Interpolazione lineare per trovare la gittata esatta a y=0
            fraction = y[-1] / (y[-1] - next_y)
            gittata_finale = x[-1] + fraction * (next_x - x[-1])
            x.append(gittata_finale)
            y.append(0.0)
            break
            
        x.append(next_x)
        y.append(next_y)
    
    gittata = x[-1]
    
    # Creazione del grafico
    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label=f'Traiettoria (m={m}kg)', color='forestgreen', linewidth=2)
    plt.axhline(0, color='black', linewidth=1) # Terreno
    
    # Disegno stilizzato del cannone (un rettangolo e un cerchio)
    plt.gca().add_patch(plt.Rectangle((-2, 0), 2, 1, color='gray'))
    plt.gca().add_patch(plt.Circle((-1, 0.5), 0.5, color='black'))
    
    plt.title(f"Simulazione Tiro del Cannone\n$v_{{0x}}={v0x} m/s, v_{{0y}}={v0y} m/s, m={m} kg$")
    plt.xlabel("Distanza (m)")
    plt.ylabel("Altezza (m)")
    plt.grid(True, linestyle=':', alpha=0.7)
    plt.legend()
    
    # Annotazione gittata
    plt.annotate(f'Gittata: {gittata:.2f} m', xy=(gittata, 0), xytext=(gittata-10, max(y)*0.2),
                 arrowprops=dict(facecolor='black', shrink=0.05),
                 horizontalalignment='right')
    
    plt.savefig('traiettoria_cannone.png')
    return gittata

# Esempio di utilizzo
v0x_val = 30
v0y_val = 40
massa_val = 12.0
gittata_risultato = simula_cannone(v0x_val, v0y_val, massa_val)

print(f"Gittata calcolata: {gittata_risultato:.2f} m")
```

# CREARE UN MODELLO CON LA MATEMATICA #

## Il Problema Teorico: L'Equilibrio della Natura
Immaginiamo un ecosistema isolato con due sole specie:

Le **Prede** (es. Alci): Hanno cibo illimitato, ma **il loro numero cala se incontrano i predatori.

I **Predatori** (es. Lupi): Muoiono di fame se non ci sono prede, ma prosperano se ne cacciano in abbondanza.

**Il quesito matematico** è: le due popolazioni raggiungeranno un equilibrio stabile, si estingueranno o oscilleranno per sempre?

# IL MODELLO MATEMATICO

Siano:

$x$: il numero di Prede, $y$: numero di Predatori

$\Delta x$ e $\Delta y$: la variazione del numero di individui nell'intervallo $\Delta t$.

## 1. Equazione delle Prede

La variazione delle prede dipende da quante ne nascono e da quante ne vengono mangiate: $$\Delta x = (\alpha \cdot x - \beta \cdot x \cdot y) \cdot \Delta t$$

$\alpha \cdot x$: rappresenta le nascite. Più prede ci sono, più ne nascono (crescita lineare).

$\beta \cdot x \cdot y$: rappresenta le morti per predazione. La probabilità di un incontro fatale dipende dal prodotto delle due popolazioni.


## 2. Equazione dei Predatori

La variazione dei predatori dipende da quanto cibo trovano e da quanti ne muoiono naturalmente:
$$\Delta y = (\delta \cdot x \cdot y - \gamma \cdot y) \cdot \Delta t$$

$\delta \cdot x \cdot y$: rappresenta le nascite. I predatori aumentano solo se mangiano prede.

$\gamma \cdot y$: rappresenta le morti naturali. Senza cibo, i predatori diminuiscono proporzionalmente al loro numero.

Tabella dei Parametri del Modello Lotka-Volterra$$\begin{array}{|c|l|l|}
\hline
\text{Parametro} & \text{Significato Biologico} & \text{Effetto sul Modello} \\ \hline
\alpha & \text{Tasso di natalità delle prede} & \text{Se alto, la popolazione di prede cresce velocemente.} \\ \hline
\beta & \text{Efficacia della caccia (mortalità)} & \text{Rappresenta la capacità dei predatori di catturare prede.} \\ \hline
\delta & \text{Efficienza nutritiva (conversione)} & \text{Quante prede servono per generare un nuovo predatore.} \\ \hline
\gamma & \text{Tasso di mortalità dei predatori} & \text{Velocità con cui i predatori diminuiscono senza cibo.} \\ \hline
\end{array}$$


# IL MODELLO PREVEDE UN SISTEMA DI DUE EQUAZIONI IN DUE INCOGNITE X E Y
(oss. x e y sono due funzioni del tempo)

# Modello **Preda-Predatore di Lotka-Volterra** (primo esempio di SISTEMA DINAMICO) :

$$\begin{cases}
\Delta x = (\alpha \cdot x - \beta \cdot x \cdot y) \cdot \Delta t \\
\Delta y = (\delta \cdot x \cdot y - \gamma \cdot y) \cdot \Delta t
\end{cases}$$

Fissati i parametri $\alpha, \beta, \delta, \gamma$ e fissato il numero iniziale di prede e predatori $x_0, y_0$
le soluzioni del sistema x(t) e y(t) mi dicono come evolvono le popolazioni nel tempo


# IL SISTEMA E' COMPLESSO...COME OTTENGO LE SOLUZIONI?

# SOLUZIONI ANALITICHE : SERVE UNA LAUREA 

# SOLUZIONI NUMERICHE CON PYTHON E GEMINI : 10 SECONDI

## Proviamo a plottare le soluzioni delle equazioni partendo da alcune condizioni iniziali

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Definizione dei parametri (coefficienti biologici)
alpha = 1.1  # Natalità prede
beta = 0.01   # Efficacia caccia
delta = 0.001  # Efficienza nutritiva
gamma = 0.4  # Mortalità predatori

# 2. Calcolo teorico dei punti di equilibrio
x_eq = gamma / delta 
y_eq = alpha / beta 

# provare con 10000,30 - 5000, 40
x_0 = 500; x_eq=x_0    # numero di prede
y_0 = 30;  y_eq=y_0    # numero di predatori

print(f"Punto di equilibrio calcolato:")
print(f"Numero di Prede (x): {x_eq}")
print(f"Numero di Predatori (y): {y_eq}")

# 3. Simulazione numerica (Metodo di Eulero)
dt = 0.01          # Passo temporale piccolo per precisione
passi = 10000      # Durata della simulazione
t = np.linspace(0, 100, passi)

# Inizializziamo le liste con i valori iniziali impostati all'equilibrio
x = np.zeros(passi)
y = np.zeros(passi)

x[0] = x_eq  # Partiamo esattamente dal valore di equilibrio
y[0] = y_eq

# Ciclo for che applica il sistema con il Delta
for i in range(passi - 1):
    delta_x = (alpha * x[i] - beta * x[i] * y[i]) * dt
    delta_y = (delta * x[i] * y[i] - gamma * y[i]) * dt

    x[i+1] = x[i] + delta_x
    y[i+1] = y[i] + delta_y

# 4. Visualizzazione
plt.figure(figsize=(10, 5))
plt.plot(t, x, label='Prede (x)', color='blue', linewidth=2)
plt.plot(t, y, label='Predatori (y)', color='red', linestyle='--', linewidth=2)
plt.ylim(0, max(x_eq, y_eq) * 1.5)
plt.title("Simulazione Lotka Volterra")
plt.xlabel("Tempo")
plt.ylabel("Popolazione")
plt.legend()
plt.grid(True)
plt.show()
```

OBIETTIVO: ottenere "L'Equilibrio Statico"

"Esiste una condizione in cui le popolazioni rimangono costanti nel tempo?

"Matematicamente, questo accade quando $\Delta x = 0$ e $\Delta y = 0$.

Escludendo il caso banale in cui le popolazioni sono estinte ($x=0, y=0$), otteniamo i punti di equilibrio dinamico:

$x_{eq} = \frac{\gamma}{\delta}$

$y_{eq} = \frac{\alpha}{\beta}$

`MODELLO MATEMATICO DI UN SISTEMA FISICO: IL MOTO DI CADUTA LIBERA`

Vediamo un altro semplicissimo modello matematico che rappresenta una situazione fisica: il moto di caduta libera senza attrito.

Scegliamo la Caduta Libera, poiché permette di vedere come una variabile (la velocità) cambi nel tempo a causa di una costante (l'accelerazione di gravità).

Il Modello Fisico

L'obiettivo è calcolare la posizione ($s(t)$) e la velocità ($v(t)$) di un oggetto che cade da un'altezza iniziale dopo un certo tempo ($t$).

Le formule fisiche che utilizzeremo sono:

Velocità: $v = g \cdot t$

Spazio percorso: $s = \frac{1}{2} \cdot g \cdot t^2$

Dove $g$ è l'accelerazione di gravità sulla Terra, circa $9.81 \, \text{m/s}^2$.
